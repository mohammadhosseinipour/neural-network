Separate input and label in train set and test set then convert them to numpy

# NOW lets Creat our model


```python
class model_creator:
    
    def __init__(self, number_layer,array_number_neron):
        self.x=tf.keras.Sequential()
        for i in range(number_layer):
            self.x.add(tf.keras.layers.Dense(array_number_neron[i], activation='relu'))    
```

we dont khow our hyperparameters so we have to use random library to find the best hyperparameters that satisfy our 
neural network.

```python
def randomise_hyperparameters():
    hyperparameters={}
    hyperparameters['number_epoch']=(50*(randrange(8)+1))
#     hyperparameters['number_epoch']=10
    
    h=randrange(2)
    h=h-4
    hyperparameters['learning_rate']=(10**h)*(randrange(9)+1)
    hyperparameters['number_layer']=randrange(5)+5
    number_layer=hyperparameters['number_layer']
    layer_neron=[]
    for i in range(number_layer):
        layer_neron.append((randrange(5)+1)*10)
    
    hyperparameters['layer_neron']=layer_neron

    return hyperparameters
    
```

we use ploting function to draw train cost and validation cost

```python
def ploting(history):    
    training_loss = history.history['loss']
    test_loss = history.history['val_loss']
    
    epoch_count = range(1, len(training_loss) + 1)
    
    plt.plot(epoch_count, training_loss, 'r--')
    plt.plot(epoch_count, test_loss, 'b-')
    plt.legend(['Training Loss', 'Test Loss'])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.show();
    
    
    
    
    training_accc = history.history['accuracy']
    test_accc = history.history['val_accuracy']
    
    epoch_count = range(1, len(training_loss) + 1)
    
    plt.plot(epoch_count, training_accc, 'r--')
    plt.plot(epoch_count, test_accc, 'b-')
    plt.legend(['Training accuracy', 'Test accuracy'])
    plt.xlabel('Epoch')
    plt.ylabel('accuracy')
    plt.show();
```

For calculating precision , recall and F_measure we have to calculate True Positive (TP), False Negative (FN), False Positive (FP) and True Negative (TN).
Precision = TruePositives / (TruePositives + FalsePositives)
Recall = TruePositives / (TruePositives + FalseNegatives)
F-Measure = (2 * Precision * Recall) / (Precision + Recall)

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    M={}
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1
    M["TP"]=TP
    M["FP"]=FP
    M["TN"]=TN
    M["FN"]=FN
    return M
```

for creating  our model we first make hidden layers and nodes with model_creator function then we add two more layers , one for detecting nodes (here we have 2 classes so we have to add a layer with 2 nodes) and a softmax node.
we train out model and test our test data set to defect the right hyperparameters.
it will trap in while loop till it satisfy our desired accuracy.

```python

train_input_copy=train_input
train_label_copy=train_label


test_input_copy=test_input
test_label_copy=test_label
array_of_models=[]
inter=0
while(1):
    train_input=train_input_copy
    train_label=train_label_copy
    
    test_input=test_input_copy
    test_label=test_label_copy
    
    
    hyperparameters=randomise_hyperparameters()
    print(hyperparameters)
    number_epoch=hyperparameters['number_epoch']    
    learning_rate=hyperparameters['learning_rate']
    number_layer=hyperparameters['number_layer']   
    layer_neron=hyperparameters['layer_neron']

    x=model_creator(number_layer,layer_neron)
    x.x.add(tf.keras.layers.Dense(2))
    x.x.add(tf.keras.layers.Softmax())
    
    array_of_models.append(x.x)
    print("rate:")
    print(learning_rate)
    print("steps:")
    print(number_epoch)

    array_of_models[inter].compile(optimizer=tf.keras.optimizers.Adam(
    learning_rate=learning_rate, amsgrad=False, name='Adam'),
                      loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                    metrics=["accuracy"])


    history=array_of_models[inter].fit(train_input,train_label, validation_data=(test_input,  test_label),epochs=number_epoch)
    
    
    array_of_models[inter].summary()
    
    train_loss, train_acc=array_of_models[inter].evaluate(train_input,train_label)
    print('train_acc:'+str(train_acc))
    test_loss, test_acc = array_of_models[inter].evaluate(test_input,  test_label, verbose=2)
    print('test_acc:'+str(test_acc))
    inter=inter+1
    
    score = x.x.evaluate(test_input, test_label, verbose=1)
    print("score:",score)
    print("raw prediction of model:\n",x.x.predict(test_input, verbose=1))
    pred_y=x.x.predict_classes(test_input, verbose=1)
    print("finial prediction of model:\n",pred_y)
    measure=perf_measure(test_label,pred_y)
    if (measure['FP']+measure['TP']) and (measure['TP']+measure['FN']):
        Precision=measure['TP']/(measure['FP']+measure['TP'])
        Recall=measure['TP']/(measure['TP']+measure['FN'])
    else:
        continue
    print("PRECISION:\n",Precision)
    print("RECALL:\n",Recall)
    F_Measure = (2 * Precision * Recall) / (Precision + Recall)
    print("F_Measure:\n",F_Measure)
    ploting(history)
    
    Difference=abs(train_loss-test_loss)
    Min=min(train_acc,test_acc)
    print('Difference:')
    print(Difference)
    print('Min:')
    print(Min)
    if(Min>=0.84 and Difference<0.03):
        print("!!!finished!!!")
        print("!!!       it\'s just Right             !!!")
        break
    else:
        if(Min<0.84):
            print('your model is under fit')
        elif(Difference>0.03 and Min>=0.84):
            print('your model is over fit')
        else:
            print('your model isn\'t good at all')
        time.sleep(5)
```
